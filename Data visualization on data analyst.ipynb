## Data visualization

### Univariate Analysis
### To understand the quality of one feature.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,8))
plt.hist(x="salary_in_usd",data = da_sal_DA)
plt.xlabel("Salary in USD")
plt.ylabel("Frequency")
plt.title("Salary distribution",size=20)
plt.show()
```

```python
plt.figure(figsize=(11,11))
sns.countplot(x="company_location",data=da_sal_DA)
plt.xlabel("Company Location")
plt.ylabel("Count")
plt.title("Company Location Comparison by count")
plt.show()
```

```python
plt.figure(figsize=(11,11))
sns.countplot(x="employee_residence",data=da_sal_DA)
plt.xlabel("Employee Residence")
plt.ylabel("Count")
plt.title("Employee Residence Comparison by count")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x="company_size",data=da_sal_DA)
plt.xlabel("Company Size")
plt.ylabel("Count")
plt.title("Company size Comparison by count")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x="remote_ratio",data=da_sal_DA)
plt.xlabel("Remote Work acceptance")
plt.ylabel("Count")
plt.title("No.of.Organizations accepts remote work by percentage")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x="experience_level",data=da_sal_DA)
plt.xlabel("Experience level of Employees")
plt.ylabel("Count")
plt.title("No.of.Employees ordered by experience level")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x="employment_type",data=da_sal_DA)
plt.xlabel("Experience level of Employees")
plt.ylabel("Count")
plt.title("No.of.Employees ordered by Job Type")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(x=da_sal_DA['company_size'].value_counts(),autopct = '%1.0f%%',labels=['M','L','S'],explode=[0.09,0,0],shadow=True)
plt.title("Piechart to demonstrate distribution between the size of the organizations")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(x=da_sal_DA['experience_level'].value_counts(),autopct = '%1.0f%%',labels=["SE","MI","EN","EX"],explode=[0.09,0,0,0],shadow=True)
plt.title("Piechart to demonstrate distribution between the experience level of the employees")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(x=da_sal_DA['remote_ratio'].value_counts(),autopct = '%1.0f%%',labels=[100,0,50],explode=[0.09,0,0],shadow=True)
plt.title("Piechart to demonstrate distribution between the remote ratio of the employees")
plt.show()
```
